```python
dial_positions = [50]
part2_ans = 0

with open('inputs/day01.txt') as f:
    for line in f.readlines():
        isleft = line[0] == 'L'
        unmod_pos = dial_positions[-1] + int(line.strip()[1:]) * (-1)**(line[0] == 'L')

        if not isleft:
            part2_ans += unmod_pos // 100
        elif dial_positions[-1] == 0:
            part2_ans += abs(unmod_pos) // 100
        else:
            part2_ans += abs((unmod_pos - 1) // 100)

        dial_positions.append(unmod_pos % 100)

print('Answer to Day 1, Part 1:', dial_positions.count(0))
print('Answer to Day 1, Part 2:', part2_ans)
```

```python
import itertools

with open('inputs/day02.txt') as f:
    ranges = f.read().split(',')
    ranges = list(map(lambda x: tuple(map(int, x.split('-'))), ranges))

def get_invalid_ids_pt1(start, end):
    if len(str(end)) > len(str(start)):
        return get_invalid_ids_pt1(start, 10 ** len(str(start)) - 1) + get_invalid_ids_pt1(10 ** len(str(start)), end)
    elif len(str(end)) % 2 == 1:
        return []
    else:
        bad_ids = [
            int(''.join(map(str, x)) + ''.join(map(str, x)))
            for x in itertools.product(range(10), repeat=int(len(str(end))/2))
            if x[0] != 0
        ]
        return list(filter(lambda x: x >= start and x <= end, bad_ids))

def get_invalid_ids_pt2(start, end):
    if len(str(end)) > len(str(start)):
        return get_invalid_ids_pt2(start, 10 ** len(str(start)) - 1) + get_invalid_ids_pt2(10 ** len(str(start)), end)
    else:
        bad_ids = []
        for factor in filter(lambda x: len(str(end)) % x == 0, range(1, len(str(end)))):
            bad_ids += [
                int(''.join(map(str, x)) * int(len(str(end)) / factor))
                for x in itertools.product(range(10), repeat=factor)
                if x[0] != 0
            ]
            
        return list(filter(lambda x: x >= start and x <= end, set(bad_ids)))

print('Answer to Day 2, Part 1:', sum(sum((get_invalid_ids_pt1(i, j) for i, j in ranges), [])))
print('Answer to Day 2, Part 2:', sum(sum((get_invalid_ids_pt2(i, j) for i, j in ranges), [])))
```

```python
with open('inputs/day03.txt') as f:
    batteries = [list(map(int, x.strip())) for x in f.readlines()]
    
    def get_max_joltage(battery_bank, n):
        if n == 1:
            return max(battery_bank)
        else:
            leading_digit = max(battery_bank[:1-n])
            return 10**(n-1) * leading_digit + \
                get_max_joltage(battery_bank[battery_bank.index(leading_digit) + 1:], n=n-1)

print('Answer to Day 3, Part 1:', sum(get_max_joltage(bb, n=2) for bb in batteries))
print('Answer to Day 3, Part 2:', sum(get_max_joltage(bb, n=12) for bb in batteries))
```

```python
import networkx as nx

g = nx.Graph()

with open('inputs/day04.txt') as f:
    for i, line in enumerate(f.readlines()):
        for j, char in enumerate(line.strip()):
            g.add_node((i, j), roll=int(char == '@'))
            if i > 0:
                g.add_edge((i, j), (i-1, j))
            if j > 0:
                g.add_edge((i, j), (i, j-1))
            if i * j > 0:
                g.add_edge((i, j), (i-1, j-1))
            if i > 0 and j < len(line.strip()) - 1:
                g.add_edge((i, j), (i-1, j+1))

part1_ans = 0
for node in g:
    if g.nodes[node]['roll'] and sum(g.nodes[n]['roll'] for n in g.neighbors(node)) < 4:
        part1_ans += 1

part2_ans = 0
while True:
    rolls_to_remove = []
    for node in g:
        if g.nodes[node]['roll'] and sum(g.nodes[n]['roll'] for n in g.neighbors(node)) < 4:
            part2_ans += 1
            rolls_to_remove.append(node)
    
    if len(rolls_to_remove) == 0:
        break

    for roll in rolls_to_remove:
        g.nodes[roll]['roll'] = 0

print('Answer to Day 4, Part 1:', part1_ans)
print('Answer to Day 4, Part 2:', part2_ans)
```

```python
import itertools

with open('inputs/day05.txt') as f:
    lines = list(f.readlines())
    ranges = list(map(lambda l: tuple(map(int, l.strip().split('-'))), lines[:lines.index('\n')]))
    ids = list(map(int, lines[lines.index('\n')+1:]))

    print('Answer to Day 5, Part 1:', sum(any(id >= r[0] and id <= r[1] for r in ranges) for id in ids))

    while True:
        for i, j in itertools.product(range(len(ranges)), repeat=2):
            if i!=j and not (ranges[i][0] > ranges[j][1] or ranges[i][1] < ranges[j][0]):
                ranges.append((min(ranges[i][0], ranges[j][0]), max(ranges[i][1], ranges[j][1])))
                del ranges[j], ranges[i] # delete bigger index first
                break
        else:
            break
    
    print('Answer to Day 5, Part 2:', sum(r[1] - r[0] + 1 for r in ranges))
```

```python
import re

with open('inputs/day06.txt') as f:
    lines = list(f.readlines())
    pt1_lines = list(map(lambda x: x.split(), lines))

part1_ans = 0
for i, j, k, l , o in zip(pt1_lines[0], pt1_lines[1], pt1_lines[2], pt1_lines[3], pt1_lines[4]):
    if o == '+':
        part1_ans += int(i) + int(j) + int(k) + int(l)
    else:
        part1_ans += int(i) * int(j) * int(k) * int(l)

print('Answer to Day 6, Part 1:', part1_ans)

problems = []
prev_match = 0
for match in re.finditer(r' (?=(\+|\*))', lines[4]):
    this_match = match.span()[0]
    problems.append((
        lines[0][prev_match: this_match],
        lines[1][prev_match: this_match],
        lines[2][prev_match: this_match],
        lines[3][prev_match: this_match],
        lines[4][prev_match: this_match].strip(),
    ))
    prev_match = this_match + 1

problems.append((
    lines[0][prev_match:-1],
    lines[1][prev_match:-1],
    lines[2][prev_match:-1],
    lines[3][prev_match:-1],
    lines[4][prev_match:].strip(),
))

part2_ans = 0
for p in problems:
    p_ans = int(p[4] == '*')
    for i, j, k, l in zip(p[0][::-1], p[1][::-1], p[2][::-1], p[3][::-1]):
        if p[4] == '+':
            p_ans += int((i + j + k + l).strip())
        else:
            p_ans *= int((i + j + k + l).strip())

    part2_ans += p_ans

print('Answer to Day 6, Part 2:', part2_ans)
```
