```python
dial_positions = [50]
part2_ans = 0

with open('inputs/day01.txt') as f:
    for line in f.readlines():
        isleft = line[0] == 'L'
        unmod_pos = dial_positions[-1] + int(line.strip()[1:]) * (-1)**(line[0] == 'L')

        if not isleft:
            part2_ans += unmod_pos // 100
        elif dial_positions[-1] == 0:
            part2_ans += abs(unmod_pos) // 100
        else:
            part2_ans += abs((unmod_pos - 1) // 100)

        dial_positions.append(unmod_pos % 100)

print('Answer to Day 1, Part 1:', dial_positions.count(0))
print('Answer to Day 1, Part 2:', part2_ans)
```
